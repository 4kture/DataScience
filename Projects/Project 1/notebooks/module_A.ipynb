```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/AmesHousing.txt', sep='\t')

df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
((df.isnull().sum() / len(df)) * 100).sort_values(ascending=False)
```

```python
df = df.drop(columns=['Pool QC', 'Misc Feature', 'Alley', 'Fence', 'Mas Vnr Type', 'Fireplace Qu'])
```

```python
df['Lot Frontage'] = df['Lot Frontage'].fillna(df['Lot Frontage'].median())
```

```python
num1 = df.select_dtypes(include='number').columns
num2 = [col for col in num1 if df[col].isnull().sum() > 0]
print(num2)
```

```python
df['Garage Yr Blt'] = df['Garage Yr Blt'].fillna(df['Year Built'])
```

```python
for col in num2:
    df[col] = df[col].fillna(df[col].median())
```

```python
df = df.fillna('Empty')
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.duplicated().sum()
```

```python
column_name1 = 'SalePrice'

plt.figure(figsize=(10, 5))
sns.histplot(df[column_name1], bins=30, kde=True, color='yellow', edgecolor='black')
plt.title(f'Гистограмма для {column_name1}')
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.grid(True)
plt.show()


plt.figure(figsize=(10, 5))
ax = sns.boxplot(x='MS Zoning', y='SalePrice', data=df, hue='MS Zoning', legend=False)  
ax.set_title("Боксплот цен продажи по зонам MS Zoning")
ax.set_xlabel("MS Zoning")
ax.set_ylabel("Цена продажи (SalePrice)")
plt.xticks(rotation=45)
plt.show()
```
