```python
import pandas as pd
import matplotlib.pyplot as plt

df_csv = pd.read_csv('data/file1.csv')
df_excel = pd.read_excel('data/file2.xlsx')
df_json = pd.read_json('data/file3.json')

df = pd.concat([df_csv, df_excel, df_json], ignore_index=True)

df
```
